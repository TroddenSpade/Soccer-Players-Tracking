```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = np.load("../masks/bg_1_trans_mask.npy")
```

```python
plt.imshow(img)
plt.show()
```

```python
_, img_b = cv2.threshold(img, 254, 255, cv2.THRESH_BINARY)
```

```python
n, C, stats, centroids = cv2.connectedComponentsWithStats(img_b)
```

```python
I_left = np.zeros(img_b.shape)
I_center = img_b.copy()
I_right = np.zeros(img_b.shape)
```

```python
I_left[C==0] = 255
I_left[:,300:] = 0
I_right[C==0] = 255
I_right[:,:300] = 0
```

```python
fig, axs = plt.subplots(3)
axs[0].imshow(I_left)
axs[1].imshow(I_center)
axs[2].imshow(I_right)
plt.show()
```

```python
np.save("../masks/I_left.npy", I_left)
np.save("../masks/I_center.npy", I_center)
np.save("../masks/I_right.npy", I_right)
```
